````python
# actual code 3
def roman_to_integer():
    romans = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}
    tot = 0
    num = input("Enter a Roman numeral: ")
    length = len(num)

    for i in range(length):
        if i + 1 < length and romans[num[i]] < romans[num[i + 1]]:
            tot -= romans[num[i]]
        else:
            tot += romans[num[i]]
    return tot
print(roman_to_integer())  # Example input: "MCMXCIV" Expected output: 1994

````

## Word Wrap at Fixed Width

**Question**

Write a function `wrap(text, width=40)` that returns a string where  
newlines (`"\\n"`) have been inserted so that no line exceeds `width`  
characters. Lines should break only at spaces between words.
```

```

````python
'''
Computational Thinking
get text from user
make empty to fill
take list and take num_chars = len(list)
for i in range(num_chars):
    new list = text[i]
    if i % 40 == 0 enter \\n
print new list

'''
````

````python
# actual code 7

def wrap(text, width=40):
    text = input("Enter text: ")
    new_text = ""
    for i in range(len(text)):
        if i % width == 0 and i != 0:
            new_text += "\n"
        new_text += text[i]

    return new_text


````

## Password Strength Checker

**Question**

Write a function `strength(pwd)` that returns one of 'weak', 'medium', or 'strong' based on these rules:
- 'strong': length >= 8 and contains at least one lowercase letter, one uppercase letter, one digit,
  and one punctuation symbol.
- 'medium': length >= 6 and contains at least one letter (upper or lower).
- otherwise: 'weak'.
```


```

````python
'''
Computational Thinking

get password input
iterate through-check length with counter
+= for upper, lower, digit for their own counters 
after iteration check if counts are > 0 or for len >= 8
return strength
'''
````

````python
# actual code 12


known_symbols = ['.', '.', '/', '<', '>', '?', ';', ':', '[', ']', '{', '}', '|', '!', '@', '#', '$', '%', '^', '&', '*', '(', ')', '-', '_', '=', '+']

def strength(pwd):
    length_counter = 0
    upper_counter = 0
    lower_counter = 0 
    number_counter = 0
    symbol_counter = 0
    for char in pwd:
        length_counter += 1
        if pwd(char).isupper():
            upper_counter += 1
        elif pwd(char).islower():
            lower_counter += 1
        elif pwd(char).isdigit():
            number_counter += 1
        elif pwd(char) in known_symbols:
            symbol_counter += 1
    if length_counter >= 8 & upper_counter >= 1 & lower_counter >= 1 & number_counter >= 1 & symbol_counter >= 1:
        return 'strong'
    elif length_counter >= 6 & (upper_counter >= 1 or lower_counter >= 1):
        return 'medium'
    else:
        return 'weak'
    
print(strength("P@ssw0rd"))  # Expected output: 'strong'

````
